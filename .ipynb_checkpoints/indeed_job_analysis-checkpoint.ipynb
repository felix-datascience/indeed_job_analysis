# Data analysis of data science jobs listed on Indeed

## Data sources

**data_science_jobs_2020-10-01.json:**

Scraped data from de.indeed.com. It consists of search results for the search phrase "data science". For each result I scraped the job title, company, location, salary and a short description. If you want to find out more about how I did that and see the code of the web scraper have a look at my repository on GitHub: https://github.com/felix-datascience/indeed_scraper

## Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading data

```python
jobs_path = './data/data_science_jobs_2020-10-01.json'
jobs_df = pd.read_json(jobs_path)
```

## Data cleaning

From the first 5 rows we can already see that there is a **unnecessary "\n"** at the beginning of each string in the **"company" column**.

The salary column contains only 45 non-null values due to most job ads not giving any information about salary.
If we take a closer look we can see that we have the same **unnecessary "\n"** at the beginning of each non-null **salary** value.

```python
jobs_df.head(10)
```

```python
jobs_df.info()
```

```python
print(type(jobs_df.salary.iloc[0]))
```

Lets remove the "\n" from the "company" and "salary" column.

```python
def remove_n(string):
    if type(string) == str:
        return string[1:]
    # return NoneType values without manipulation
    else:
        return string

jobs_df['company'] = jobs_df['company'].apply(remove_n)
jobs_df['salary'] = jobs_df['salary'].apply(remove_n)

jobs_df.head(10)
```

```python
# 30 most common cities
most_common_cities = jobs_df['location'].value_counts()[:20]

plt.figure(figsize=(30,20))

most_common_cities_fig = sns.barplot(
    x=most_common_cities.index.values,
    y=most_common_cities.values
)

most_common_cities_fig.set_xticklabels(
    most_common_cities_fig.get_xticklabels(),
    rotation=45,
    fontsize='xx-large',
    fontweight='bold'
)

most_common_cities_fig.set_yticklabels(
    most_common_cities_fig.get_yticklabels(),
    fontsize='xx-large',
    fontweight='bold'
)
```
